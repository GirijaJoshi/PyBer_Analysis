```python
# Import dependencies.
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
# Load in csv
file_name = os.path.join(".", "Resources", "PyBer_ride_data.csv")
pyber_ride_df = pd.read_csv(file_name)
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
print(x_axis)
tick_locations = [value for value in x_axis]
print(tick_locations)
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations,pyber_ride_df["Month"] )
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
# plt.xticks(tick_locations,pyber_ride_df["Month"])
plt.show()
```

```python
# Add Error
import statistics
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
stdev
```

```python
# Change the color of the bars to sky blue.Add Cap size. 
# Rotate the labels on the x-axis to horizontal. Set the y-axis increment to every $5.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar", 
                   yerr=stdev, capsize=3, color="skyblue", )
print(pyber_ride_df["Avg. Fare ($USD)"].max())
print(pyber_ride_df["Avg. Fare ($USD)"].min())
y_ticks = np.arange(0, pyber_ride_df["Avg. Fare ($USD)"].max() + 10, step=5.0)
print(y_ticks)
plt.yticks(y_ticks)
plt.xticks(rotation=0)
```
