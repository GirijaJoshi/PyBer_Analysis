```python
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# object-oriented approach.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# create a bubble chart using the object-oriented interface approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# using the object-oriented approach, make the following changes to your bubble chart:
# Change the color of the markers to sky blue. Change the size of the markers to 5 times each data point.
# Make the color 20% transparent. Add a black edge color to the circles. #Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.

fig, ax = plt.subplots()
# ax.scatter(y_axis, x_axis, c='skyblue', edgecolors='black', s = [i * 5 for i in y_axis], label="Boston")
ax.scatter(y_axis, x_axis, c='skyblue', edgecolors='black', s = [i * 5 for i in y_axis], 
           linewidth=1, label="Boston", alpha=0.8)
# Add a title and axes labels.
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Invert Y Axis
ax.invert_yaxis()
# Create a title.
plt.title("Pyber Fare by Month")
ax.legend()
```
