```python
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# To whitch line to scatter we need to add 'o'
plt.plot(x_axis, y_axis, 'o')
```

```python
# There is another way to create scatter plot
plt.scatter(x_axis, y_axis)
```

```python
# There is another way to create scatter plot. Change the color of the markers to red. Add a legend for the city of Chicago.
# Switch the axis so that the Fare($) data is on the x-axis.
plt.scatter(y_axis, x_axis, c="red",label="Chicago" )
# Add a title and axes labels.
plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.gca().invert_yaxis()
# Add a limit for the x-axis data.
plt.xlim(0, 50)
plt.legend()
```

```python
# Bubble charts are useful when presenting financial, population, and weather data 
# because you can add a third and fourth factor to convey more information.
# By changing the “dot” into a “bubble,” we are adding a third factor: size. 
# If there is more than one dataset that uses the same axes, we can change the color of each marker for each dataset, 
# which will add a fourth factor: color.

# marker size has changed based on the value of the fare in USD, as you can see here
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# We can adjust the size by multiplying the data in the y-axis by any value.
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Python way to do same as above
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# User random color for each marter. N should be same Max value on X axis
N = 12
colors = np.random.rand(N)
plt.scatter(x_axis, y_axis, c=colors, s = [i * 3 for i in y_axis])
```
