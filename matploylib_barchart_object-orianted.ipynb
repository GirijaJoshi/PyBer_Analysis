```python
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# To create a bar chart using the object-oriented interface method, use the ax.bar()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
# ax.bar(x_axis, y_axis, color="black", label='New York')
ax.bar(x_axis, y_axis, color="k", label='New York')
ax.legend()
```

```python
# To create a horizontal bar chart using the object-oriented interface
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# If we want the data on opposite axes
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# To create a horizontal bar chart using the object-oriented interface
# Change the bar color to cyan.Add a legend for the city of Chicago.
# Switch the axis so that the Fare($) data is on the x-axis.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
# Invert Y Axis
ax.invert_yaxis()
# Add a title and axes labels.
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
# Add a title and axes labels.
plt.title("Pyber Fare by Month")

ax.legend()
```
